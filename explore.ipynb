```python
#Imports
import numpy as np
import pandas as pd
```

```python
#Load player dataframe
player = pd.read_csv('./lahman/player.csv')
player.head()
```

```python
#Load batting dataframe
batting = pd.read_csv('./lahman/batting.csv')
batting.head()
```

```python
player_batting = player.merge(batting, how='inner', on='player_id')
player_batting.head()
```

```python
# Add calculated field avg, slg
player_batting['avg'] = player_batting['h']/player_batting['ab']
player_batting['slg'] = (
    player_batting['h'] - player_batting['double'] - player_batting['triple'] - player_batting['hr']  # Singles
    + 2*player_batting['double'] 
    + 3*player_batting['triple'] 
    + 4*player_batting['hr']
) / player_batting['ab']
```
