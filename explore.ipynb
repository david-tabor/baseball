```python
#Imports
import numpy as np
import pandas as pd
```

```python
#Load player dataframe
player = pd.read_csv('./lahman/player.csv')
player.head()
```

```python
#Load batting dataframe
batting = pd.read_csv('./lahman/batting.csv')
batting.head()
```

```python
player_batting = player.merge(batting, how='inner', on='player_id')
player_batting.head()
```

```python
# Find Mike Trout

mtrout = player_batting[player_batting['player_id'].str.contains('trout')]
mtrout.sort_values('birth_year', ascending=False)

mtrout = player_batting.loc[lambda df: df['player_id'] == 'troutmi01']
mtrout
```

```python
list(mtrout)

traditional = ['name_first', 'name_last', 'year', 'team_id', 'g', 'ab', 'r', 'h', 'hr', 'rbi', 'sb', 'bb', 'so']
mtrout[traditional].sort_values('year')
```

```python

list(mtrout)

mtrout['avg'] = mtrout['h']/mtrout['ab']
mtrout['avg']

singles = mtrout['h'] - mtrout['double'] - mtrout['triple'] - mtrout['hr']
mtrout['slg'] = (singles + 2*mtrout['double'] + 3*mtrout['triple'] + 4*mtrout['hr']) / mtrout['ab']
mtrout['slg']
```
